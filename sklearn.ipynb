```python
import pandas as pd
import matplotlib.pyplot as plt

xlsx = pd.ExcelFile('DataForPerceptron.xlsx')
train_data = pd.read_excel(xlsx, 'TRAINData', index_col=None)
test_data = pd.read_excel(xlsx, 'TESTData', index_col=None)
```

```python
x_train = train_data.drop(['SubjectID', 'Class'], axis=1)
y_train = train_data['Class']

x_test = test_data.drop(['SubjectID', 'Class'], axis=1)
```

```python
X = x_train
y = y_train
```

**---------------**

```python
from sklearn.linear_model import Perceptron
```

```python
model = Perceptron()
```

```python
print(X.shape, y.shape)
```

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
res = []

for values in x_train.values:
    res.append(model.predict([values])[0])
```

```python
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold

cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
# evaluate model
scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1)
# summarize result
print('Mean Accuracy: %.3f (%.3f)' % (mean(scores), std(scores)))
```
